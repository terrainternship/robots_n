```python
# librosa

import librosa
import IPython
import matplotlib.pyplot as plt
import librosa.display as ld

signal, sr = librosa.load('/content/drive/MyDrive/Colab Notebooks/УИИ/Стажировка | Tetrabot/Присланные команды/Голос близко/Включай автоматический режим движения.WAV')
```

```python
print(signal.shape, sr)
```

```python
print(signal)
```

```python
plt.figure(figsize=(12,4))
ld.waveshow(signal, sr=sr)
```

```python
display(IPython.display.Audio(signal, rate = sr))
```

```python
import numpy as np

n_fft = 2048
ft = np.abs(librosa.stft(signal[:n_fft], hop_length = n_fft+1))
plt.plot(ft)
plt.title('Spectrum')
plt.xlabel('Frequency Bin')
plt.ylabel('Amplitude')
```
